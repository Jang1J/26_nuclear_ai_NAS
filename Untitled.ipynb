## 6. Results Comparison

```python
import os

print("=" * 70)
print(f"{'Model':<55s} {'Acc':>6s}  {'Loss':>8s}")
print("=" * 70)

test_dir = "test_results"
for run_name in sorted(os.listdir(test_dir)):
    metrics_file = os.path.join(test_dir, run_name, "test_metrics.txt")
    if not os.path.isfile(metrics_file):
        continue
    with open(metrics_file) as f:
        lines = f.readlines()
    acc_line = [l for l in lines if l.startswith("test_acc")]
    loss_line = [l for l in lines if l.startswith("test_loss")]
    acc = acc_line[0].split(":")[1].strip() if acc_line else "N/A"
    loss = loss_line[0].split(":")[1].strip() if loss_line else "N/A"
    print(f"{run_name:<55s} {acc:>6s}  {loss:>8s}")

print("=" * 70)
```
